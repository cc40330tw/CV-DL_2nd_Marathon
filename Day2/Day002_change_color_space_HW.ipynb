使用 [Lena](http://optipng.sourceforge.net/pngtech/img/lena.png)

- 改變圖片的 color space (`HLS`, `LAB`) 並呈現

```python
import cv2
imagePath = '/CV+DL_2nd_Marathon/lena.png'
image = cv2.imread(imagePath)
image_HLS = cv2.cvtColor(image,cv2.COLOR_BGR2HLS)
image_LAB = cv2.cvtColor(image,cv2.COLOR_BGR2LAB)

while True: #就算把叉叉
    cv2.imshow('Lena',image)
    cv2.imshow('Lena_HLS',image_HLS)
    cv2.imshow('Lena_LAB',image_LAB)
    k = cv2.waitKey(0)
    if k == 27:
        cv2.destroyAllWindows()
        break
```

```python
#以下為自行練習
import cv2
imagePath = 'cat.jpg'
image = cv2.imread(imagePath)
image_HLS = cv2.cvtColor(image,cv2.COLOR_BGR2HLS)
image_LAB = cv2.cvtColor(image,cv2.COLOR_BGR2LAB)

cv2.namedWindow("Cat",cv2.WINDOW_NORMAL)
cv2.imshow("Cat",image)
cv2.namedWindow("Cat_HLS",cv2.WINDOW_NORMAL)
cv2.imshow("Cat_HLS",image_HLS)
cv2.namedWindow("Cat_LAB",cv2.WINDOW_NORMAL)
cv2.imshow("Cat_LAB",image_LAB)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
imagePath = 'cat.jpg'
image_BGR = cv2.imread(imagePath,cv2.IMREAD_COLOR)
image_hsv = cv2.cvtColor(image_BGR,cv2.COLOR_BGR2HSV)

cv2.namedWindow("Cat",cv2.WINDOW_NORMAL)
cv2.imshow("Cat",image_BGR)
cv2.namedWindow("Cat_HSV",cv2.WINDOW_NORMAL)
cv2.imshow("Cat_HSV",image_hsv)

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
from matplotlib import pyplot as plt

imagePath = 'cat.jpg'
image_BGR = cv2.imread(imagePath,cv2.IMREAD_COLOR)
image_RGB = cv2.cvtColor(image_BGR,cv2.COLOR_BGR2RGB)

cv2.namedWindow("Cat_BGR",cv2.WINDOW_NORMAL)
cv2.imshow("Cat_BGR",image_BGR) #OpenCV把圖讀進來是以BGR的方式儲存

cv2.namedWindow("Cat_RGB",cv2.WINDOW_NORMAL)
cv2.imshow("Cat_RGB",image_RGB)

cv2.waitKey(0)
cv2.destroyAllWindows()

#====================================================================================
plt.imshow(image_BGR) #如果把OpenCV讀入的圖直接以Matplotlib顯示，就會讀出BGR而非原圖的RGB
plt.show()
plt.imshow(image_RGB)
plt.show()
```
